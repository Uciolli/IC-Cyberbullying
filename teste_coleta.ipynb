```python
import pandas as pd
lexicon = pd.read_csv('csvs/lexicon_enriquecido.csv')
```

```python
import time
results = []
user_info = []
for i in range(len(lexicon)):
    print(i)
    if i == 200 or i == 400 or i == 600 or i == 800:
        print('Dormindo...')
        time.sleep(1200)
        print('Acordei!')
    tweets = client.search_recent_tweets(query=lexicon.loc[i, 'Portuguese (original)'], max_results=10, expansions=['author_id'])
    tweets_data = tweets.data
    users = tweets.includes
    if not tweets_data is None and len(tweets_data) > 0:
        for tweet in tweets_data:
            obj ={}
            obj['id'] = tweet.id
            obj['text'] = tweet.text
            results.append(obj)
        for user in users['users']:
            obj={}
            obj['user_id'] = user.id
            user_info.append(obj)
```
