```python
import tweepy as tw
client = tw.Client(
                   bearer_token        = '',
                   consumer_key        = '',
                   consumer_secret     ='',
                   access_token        = '',
                   access_token_secret =''
                   )
```

```python
tweets = client.search_recent_tweets(query='Lula', max_results=10, expansions=['author_id'])
```

```python
terms = ['Lula', 'luladr√£o', 'lularapio', 'ladr√£o', 'Lulinha']
#terms = ['Bolsonaro', 'bolsonazi', 'bozo', 'bolsomerda', 'bolsomito']
```

```python
import time
results = []
user_info = []
j = 0
for term in terms:
    if j == 200 :
        j = 0
        print('Dormindo...')
        time.sleep(300)
        print('Acordei!')
    tweets = client.search_recent_tweets(query= term +' -is:retweet', max_results=100, expansions=['author_id'])
    tweets_data = tweets.data
    users = tweets.includes
    if not tweets_data is None and len(tweets_data) > 0:
        for tweet in tweets_data:
            obj ={}
            obj['id'] = tweet.id
            obj['text'] = tweet.text
            results.append(obj)
        for user in users['users']:
            obj={}
            obj['user_id'] = user.id
            user_info.append(obj)
    j+=1
```

```python
import pandas as pd
df1 = pd.DataFrame(results, columns=['id','text'])
df2 = pd.DataFrame(user_info, columns=['user_id'])
df1=df1.applymap(str)
df2=df2.applymap(str)
dfs = pd.concat([df1,df2],axis=1)
dfs = dfs.drop_duplicates()
dfs.count()
```
